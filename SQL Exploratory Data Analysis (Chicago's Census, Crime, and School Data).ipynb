# 2. Exploratory Data Analysis with SQL and Pandas

### Total number of crimes recorded in the CRIME table.

```python
%%sql
SELECT COUNT(*) AS TOTAL_CRIME_RECORDED 
FROM CHICAGO_CRIME_DATA
```

### Total number of crimes recorded in All Area Through the Year

```python
data = pd.read_sql_query('''
SELECT CCD.YEAR, COUNT(*) AS TOTAL_CRIME
FROM CHICAGO_CRIME_DATA AS CCD
JOIN CENSUS_DATA AS CD
ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
GROUP BY CCD.YEAR
''', con)
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['YEAR'], df['TOTAL_CRIME'])
plt.title('Total Crime in All Area Through the Year')
plt.xlabel('Year')
plt.ylabel('Total Crime')
plt.xticks(df['YEAR'], df['YEAR'])
plt.show()
```

### Top 5 Crime Type in All Area

```python
data = pd.read_sql_query('''
SELECT PRIMARY_TYPE ,COUNT(*) AS TOTAL_CRIME_TYPE
FROM CHICAGO_CRIME_DATA
GROUP BY PRIMARY_TYPE
ORDER BY TOTAL_CRIME_TYPE DESC
LIMIT 5''',con)
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(25,10))
plt.bar(df['PRIMARY_TYPE'], df['TOTAL_CRIME_TYPE'])
plt.title('Top 5 Crime Type in All Area')
plt.xlabel('Crime Type')
plt.ylabel('Total Crime')
plt.show()
```

### Top 5 Crime Type in All Area Through the Year

```python
data = pd.read_sql_query('''
SELECT PRIMARY_TYPE, YEAR, COUNT(*) AS TOTAL_CRIME_TYPE
FROM CHICAGO_CRIME_DATA
GROUP BY PRIMARY_TYPE, YEAR
ORDER BY YEAR
''',con)
df = pd.DataFrame(data)
df
```

```python
theft=df[df['PRIMARY_TYPE']=='THEFT']
battery=df[df['PRIMARY_TYPE']=='BATTERY']
criminal_damage=df[df['PRIMARY_TYPE']=='CRIMINAL DAMAGE']
narcotics=df[df['PRIMARY_TYPE']=='NARCOTICS']
other_offense=df[df['PRIMARY_TYPE']=='OTHER OFFENSE']
```

```python
plt.figure(figsize=(20,10))
plt.plot(theft['YEAR'], theft['TOTAL_CRIME_TYPE'], label = "Theft")
plt.plot(battery['YEAR'], battery['TOTAL_CRIME_TYPE'], label = "Battery")
plt.plot(criminal_damage['YEAR'], criminal_damage['TOTAL_CRIME_TYPE'], label = "Criminal Damage")
plt.plot(narcotics['YEAR'], narcotics['TOTAL_CRIME_TYPE'], label = "Narcotics")
plt.plot(other_offense['YEAR'], other_offense['TOTAL_CRIME_TYPE'], label = "Other Offense")
plt.title('Top 5 Crime Type in All Area Through the Year')
plt.xticks(df['YEAR'], df['YEAR'])
plt.legend()
plt.show()
```

### Top 5 Crime Location in All Area

```python
data = pd.read_sql_query('''
SELECT CCD.LOCATION_DESCRIPTION ,COUNT(*) AS TOTAL_CRIME_LOCATION
FROM CHICAGO_CRIME_DATA AS CCD
JOIN CENSUS_DATA AS CD
ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
GROUP BY LOCATION_DESCRIPTION
ORDER BY TOTAL_CRIME_LOCATION DESC
LIMIT 5''',con)
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(25,10))
plt.bar(df['LOCATION_DESCRIPTION'], df['TOTAL_CRIME_LOCATION'])
plt.title('Top 5 Crime Location in All Area')
plt.xlabel('Location')
plt.ylabel('Total Crime')
plt.show()
```

### Arrests Percentage in All Area

```python
data = pd.read_sql_query('''
SELECT COUNT(*) AS TOTAL_CRIME, ARREST
FROM CHICAGO_CRIME_DATA AS CCD
JOIN CENSUS_DATA AS CD
ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
GROUP BY ARREST''',con)
df = pd.DataFrame(data)
df
```

```python
label=['Not Arrested','Arrested']
plt.pie(df['TOTAL_CRIME'], labels = label, autopct='%1.1f%%')
plt.title('Arrests Percentage in All Area')
plt.show() 
```

### Domestic Crime Percentage in All Area

```python
data = pd.read_sql_query('''
SELECT COUNT(*) AS TOTAL_CRIME, ARREST
FROM CHICAGO_CRIME_DATA
GROUP BY DOMESTIC''',con)
df = pd.DataFrame(data)
df
```

```python
label=['Not Domestic','Domestic']
plt.pie(df['TOTAL_CRIME'], labels = label, autopct='%1.1f%%')
plt.title('Domestic Crime Percentage in All Area')
plt.show() 
```

### Top 5 areas with the most crimes recorded.

```python
data = pd.read_sql_query('''
SELECT CD.COMMUNITY_AREA_NAME, COUNT(*) AS TOTAL_CRIME
FROM CHICAGO_CRIME_DATA AS CCD
JOIN CENSUS_DATA AS CD
ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
GROUP BY CD.COMMUNITY_AREA_NAME
ORDER BY TOTAL_CRIME DESC
LIMIT 5''', con)
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['COMMUNITY_AREA_NAME'], df['TOTAL_CRIME'])
plt.title('Top 5 Community Area Name with the Most Crimes Recorded')
plt.xlabel('Area Name')
plt.ylabel('Total Crime')
plt.show()
```

### Average Safety Score in Elementary, Middle, or High School

```python
%%sql
SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) AS AVG_SAFETY_SCORE 
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY "Elementary, Middle, or High School"
```

### Top 5 Area with the highest Average School Safety Score

```python
%%sql
SELECT COMMUNITY_AREA_NAME, AVG(SAFETY_SCORE) AS AVG_SAFETY_SCORE 
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY AVG_SAFETY_SCORE DESC
LIMIT 5
```

### Top 5 area with the highest college enrollment

```python
%%sql
SELECT COMMUNITY_AREA_NAME, COLLEGE_ENROLLMENT
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY COLLEGE_ENROLLMENT DESC
LIMIT 5
```

### Top 5 Area with the lowest graduation rate

```python
%%sql
SELECT COMMUNITY_AREA_NAME, Graduation_Rate__
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY Graduation_Rate__
LIMIT 5
```

### Top 5 Area with the highest Environment Score

```python
%%sql
SELECT COMMUNITY_AREA_NAME, Environment_Score
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY Environment_Score DESC
LIMIT 5
```

### Top 5 Area with the lowest average student ettendance

```python
%%sql
SELECT COMMUNITY_AREA_NAME, AVG(AVERAGE_STUDENT_ATTENDANCE) AS AVG_STUDENT_ATTENDANCE 
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY AVG_STUDENT_ATTENDANCE
LIMIT 5
```

##### Crimes involving minors / child

```python
%%sql
SELECT CASE_NUMBER, DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE DESCRIPTION LIKE '%MINOR%' OR DESCRIPTION LIKE '%CHILD%'
```

##### List all kidnapping crimes involving a child

```python
%%sql
SELECT PRIMARY_TYPE, DESCRIPTION 
FROM CHICAGO_CRIME_DATA 
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%'
```
